### Input

Some cells have input. In VSCode, this will cause a prompt to appear at the TOP of the screen for you type something and hit `Enter`. What you type is then sent to the cell. Try this one.

```python
# A block of code with input
name = input("Enter your name: ")
print(f"Hi, {name}!")
```

If you run it again and enter something different, the output will change. Try it!

### Memory

Each of the cells in a notebook shares memory with the others once it's been run. Consider what happens if you run this one.

```python
# Shared memory
print(f"Your name last time was {name}")
```

Hmm. **If** you ran the previous cell, your name has been stored, and you didn't need to enter it again.
